```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import yfinance as yf
from pandas_datareader import data as wb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.layers import GRU, Embedding, LSTM
from sklearn.preprocessing import MinMaxScaler
```

```python
tickers_list = ['^NSEI', 'DOW','^IXIC', '^N225']
stock_data = pd.DataFrame(columns=tickers_list)
for i in tickers_list:
    stock_data[i] = yf.download(i, '2007-7-1', '2009-7-1')['Adj Close']
((stock_data.pct_change()+1).cumprod()).plot(figsize=(10, 7))
plt.legend()
plt.title("Adjusted Close Price", fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.show()
```
